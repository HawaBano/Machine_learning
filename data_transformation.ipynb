```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create non normal data
np.random.seed(0)
df = np.random.exponential(size=1000, scale=2)
df = pd.DataFrame(df, columns=['values'])
df.head()
df.shape
```

```python
sns.histplot(df['values'], kde=True)
```

Methods Of Transfarmer   

1: Using BoxCox                                                  

2: Yeo_jhonson

```python
# import library
from sklearn.preprocessing import PowerTransformer
# Using boxcox 
pt_boxcox = PowerTransformer(method='box-cox')
df['box-cox'] = pt_boxcox.fit_transform(df[['values']] + 1)
```

```python
# import library
from sklearn.preprocessing import PowerTransformer
# Using yoe jhonson 
pt_yeo_johnson = PowerTransformer(method='yeo-johnson')
df['yeo_johnson'] = pt_yeo_johnson.fit_transform(df[['values']])
```

```python
df.head()
```

```python
for col in df:
    sns.histplot(df[col], kde=True)
    plt.show()
```

Normalization

1-L1 Normalization

2-L2 Normalization


```python
from sklearn.preprocessing import Normalizer
df = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
normalizer = Normalizer(norm='l2')
print(normalizer.fit_transform(df))
```

```python
from sklearn.preprocessing import Normalizer
df = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
normalizer = Normalizer(norm='l1')
print(normalizer.fit_transform(df))
```
